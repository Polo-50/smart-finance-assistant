### Foundation Data Processing Functions

````python
import io # Added this import

# The entire module content was incorrectly nested inside `load_and_clean_transaction_data`.
# This part of the code is now at the global scope of the cell.
"""
Smart Finance Assistant
Analyzes spending data from a CSV file and generates business-friendly insights.
"""

from __future__ import annotations

import logging
from dataclasses import dataclass
from pathlib import Path

import pandas as pd

logging.basicConfig(level=logging.INFO, format="%(levelname)s: %(message)s")
logger = logging.getLogger(__name__)

REQUIRED_COLUMNS = {"Date", "Amount", "Category", "Description"}


# ---------------------------------------------------------------------------
# Configuration & result types
# ---------------------------------------------------------------------------

@dataclass
class AnalysisConfig:
    top_n: int = 3
    currency: str = "AUD"
    drop_zero_amounts: bool = True


@dataclass
class SpendingInsights:
    total_spending: float
    total_refunds: float
    net_spending: float
    avg_transaction: float
    transaction_count: int
    spending_count: int
    refund_count: int
    top_categories: dict[str, float]
    category_totals: dict[str, float]
    monthly_totals: dict[str, float]
    largest_transaction: dict
    pct_by_category: dict[str, float]


# ---------------------------------------------------------------------------
# Loading & cleaning
# ---------------------------------------------------------------------------

def load_csv(csv_file: str | Path | io.StringIO) -> pd.DataFrame:
    """Load and validate the CSV file or StringIO object."""
    if isinstance(csv_file, io.StringIO):
        df = pd.read_csv(csv_file)
        if df.empty:
            raise ValueError("CSV data in StringIO is empty.")
    else:
        path = Path(csv_file)
        if not path.exists():
            raise FileNotFoundError(f"File not found: {path}")
        if path.stat().st_size == 0:
            raise ValueError("CSV file is empty.")
        df = pd.read_csv(path)

    missing = REQUIRED_COLUMNS - set(df.columns)
    if missing:
        raise ValueError(f"Missing required column(s): {', '.join(sorted(missing))}")

    return df


def clean_data(df: pd.DataFrame, config: AnalysisConfig) -> pd.DataFrame:
    """Clean and normalise raw spending data."""
    df = df.copy()

    # Normalise Amount: strip currency symbols/commas, coerce to float
    df["Amount"] = (
        df["Amount"]
        .astype(str)
        .str.replace(r"[$,]", "", regex=True)
        .str.strip()
    )
    df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

    # Parse dates
    df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

    # Drop rows that are unusable
    before = len(df)
    df = df.dropna(subset=["Amount", "Category", "Date"])
    if config.drop_zero_amounts:
        df = df[df["Amount"] != 0]
    dropped = before - len(df)
    if dropped:
        logger.warning("Dropped %d row(s) due to missing/invalid data.", dropped)

    # Normalise category casing
    df["Category"] = df["Category"].astype(str).str.strip().str.title()

    # Derived column for monthly grouping
    df["YearMonth"] = df["Date"].dt.to_period("M").astype(str)

    return df


# ---------------------------------------------------------------------------
# Metric computation
# ---------------------------------------------------------------------------

def compute_insights(df: pd.DataFrame, config: AnalysisConfig) -> SpendingInsights:
    """Derive all business metrics from clean data."""
    spending = df[df["Amount"] > 0]
    refunds  = df[df["Amount"] < 0]

    total_spending = spending["Amount"].sum()
    total_refunds  = refunds["Amount"].sum()   # negative value
    net_spending   = total_spending + total_refunds

    category_totals = (
        spending.groupby("Category")["Amount"]
        .sum()
        .sort_values(ascending=False)
    )

    pct_by_category = {
        cat: round(val / total_spending * 100, 1) if total_spending else 0.0
        for cat, val in category_totals.items()
    }

    monthly_totals = (
        spending.groupby("YearMonth")["Amount"]
        .sum()
        .to_dict()
    )

    if not spending.empty:
        idx = spending["Amount"].idxmax()
        largest = {
            "description": spending.loc[idx, "Description"],
            "amount":      spending.loc[idx, "Amount"],
            "category":    spending.loc[idx, "Category"],
            "date":        str(spending.loc[idx, "Date"].date()),
        }
    else:
        largest = {}

    return SpendingInsights(
        total_spending=total_spending,
        total_refunds=total_refunds,
        net_spending=net_spending,
        avg_transaction=df["Amount"].mean(),
        transaction_count=len(df),
        spending_count=len(spending),
        refund_count=len(refunds),
        top_categories=category_totals.head(config.top_n).to_dict(),
        category_totals=category_totals.to_dict(),
        monthly_totals=monthly_totals,
        largest_transaction=largest,
        pct_by_category=pct_by_category,
    )


# ---------------------------------------------------------------------------
# Reporting
# ---------------------------------------------------------------------------

def _fmt(amount: float, currency: str) -> str:
    return f"{currency} ${amount:,.2f}"


def print_report(insights: SpendingInsights, config: AnalysisConfig) -> None:
    """Print a formatted summary to stdout."""
    c   = config.currency
    sep = "=" * 46

    print(f"\n📊 Smart Finance Analysis")
    print(sep)
    print(f"  Gross Spending   : {_fmt(insights.total_spending,  c)}")
    print(f"  Refunds          : {_fmt(insights.total_refunds,   c)}")
    print(f"  Net Spending     : {_fmt(insights.net_spending,    c)}")
    print(f"  Avg Transaction  : {_fmt(insights.avg_transaction, c)}")
    print(f"  Transactions     : {insights.transaction_count} "
          f"({insights.spending_count} charges / {insights.refund_count} refunds)")

    print(f"\n📂 Top {config.top_n} Categories")
    print("-" * 46)
    for cat, val in insights.top_categories.items():
        pct = insights.pct_by_category.get(cat, 0)
        bar = "█" * int(pct / 2)
        print(f"  {cat:<20} {_fmt(val, c):>14}  {pct:5.1f}%  {bar}")

    if insights.monthly_totals:
        print(f"\n📅 Monthly Breakdown")
        print("-" * 46)
        for month, total in sorted(insights.monthly_totals.items()):
            print(f"  {month}   {_fmt(total, c)}")

    if insights.largest_transaction:
        lt = insights.largest_transaction
        print(f"\n💸 Largest Transaction")
        print("-" * 46)
        print(f"  {lt['description']} — {_fmt(lt['amount'], c)} "
              f"({lt['category']}, {lt['date']})")

    print(sep)


def to_dict(insights: SpendingInsights, config: AnalysisConfig) -> dict:
    """Return a JSON-serialisable dict (useful for APIs and tests)."""
    return {
        "currency":            config.currency,
        "total_spending":      round(insights.total_spending,  2),
        "total_refunds":       round(insights.total_refunds,   2),
        "net_spending":        round(insights.net_spending,    2),
        "avg_transaction":     round(insights.avg_transaction, 2),
        "transaction_count":   insights.transaction_count,
        "spending_count":      insights.spending_count,
        "refund_count":        insights.refund_count,
        "top_categories":      {k: round(v, 2) for k, v in insights.top_categories.items()},
        "category_totals":     {k: round(v, 2) for k, v in insights.category_totals.items()},
        "monthly_totals":      {k: round(v, 2) for k, v in insights.monthly_totals.items()},
        "pct_by_category":     insights.pct_by_category,
        "largest_transaction": insights.largest_transaction,
    }


# ---------------------------------------------------------------------------
# Public API
# ---------------------------------------------------------------------------

def analyze_spending_data(
    csv_file: str | Path | io.StringIO, # Modified signature to accept StringIO
    config: AnalysisConfig | None = None,
    *,
    print_summary: bool = True,
) -> dict | None:
    """
    Analyse spending data from a CSV file.

    Parameters
    ----------
    csv_file      : Path to the CSV file or a StringIO object.
    config        : Optional AnalysisConfig (uses defaults if omitted).
    print_summary : Set False to suppress stdout output.

    Returns
    -------
    Plain dict of insights, or None on failure.
    """
    if config is None:
        config = AnalysisConfig()

    try:
        df = load_csv(csv_file)
        df = clean_data(df, config)
        insights = compute_insights(df, config)

        if print_summary:
            print_report(insights, config)

        return to_dict(insights, config)

    except FileNotFoundError as e:
        logger.error("File error: %s", e)
    except ValueError as e:
        logger.error("Data error: %s", e)
    except Exception as e:
        logger.exception("Unexpected error: %s", e)

    return None


# ---------------------------------------------------------------------------
# CLI
# ---------------------------------------------------------------------------

# The `if __name__ == "__main__":` block is commented out as it's for
# command-line execution and causes errors in interactive notebooks.
# if __name__ == "__main__":
#     import argparse, json

#     parser = argparse.ArgumentParser(description="Smart Finance Analyzer")
#     parser.add_argument("csv_file",    help="Path to spending CSV")
#     parser.add_argument("--top-n",     type=int, default=3,     help="Top N categories")
#     parser.add_argument("--currency",  type=str, default="AUD", help="Currency label")
#     parser.add_argument("--json",      action="store_true",     help="Emit raw JSON")
#     args = parser.parse_args()

#     cfg    = AnalysisConfig(top_n=args.top_n, currency=args.currency)
#     result = analyze_spending_data(args.csv_file, cfg, print_summary=not args.json)

#     if args.json and result:
#         print(json.dumps(result, indent=2))

# Test your function with the sample DataFrame available in the notebook state.
# First, convert the DataFrame to a CSV string in a StringIO object.
df_sample_csv = io.StringIO(df_sample.to_csv(index=False))
analysis_result = analyze_spending_data(df_sample_csv)

print("\nAnalysis Result (dict):")
print(analysis_result)
````

````python
# 🤖 AI Collaboration: Spending Analysis Function
# Ask AI to help you create comprehensive spending analysis

def analyze_spending_patterns(df):
    """
    Analyze spending patterns and generate business insights

    🤖 AI Collaboration Prompt:
    "Create a function that analyzes spending by category, calculates percentages,
    identifies top spending areas, and generates actionable financial insights
    formatted for business presentation."

    Args:
        df: Cleaned transaction DataFrame
    Returns:
        dict: Analysis results and insights
    """
    # ── Cell 1: Imports ──────────────────────────────────────────
import io
import re
import datetime
from dataclasses import dataclass, field

import pandas as pd
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output


# ── Cell 2: Constants & thresholds ───────────────────────────

THRESHOLDS = {
    "high_concentration":  40.0,   # % — one category dominates
    "well_distributed":    20.0,   # % — top category is modest
    "low_spend":            5.0,   # % — potentially underfunded
    "spike_vs_mean":        2.0,   # multiplier — category is an outlier
    "high_single_tx":     500.0,   # currency units
}

REQUIRED_COLUMNS = {"Category", "Amount"}


# ── Cell 3: Result dataclass ─────────────────────────────────

@dataclass
class PatternInsight:
    level: str    # "warning" | "success" | "info"
    title: str
    detail: str


@dataclass
class SpendingPatternResult:
    total_spending: float
    category_totals: dict[str, float]
    category_percentages: dict[str, float]
    category_tx_counts: dict[str, int]
    category_avg_tx: dict[str, float]
    top_category: dict                    # name, amount, percentage
    outlier_categories: list[str]
    low_spend_categories: list[str]
    high_spend_categories: list[str]
    monthly_breakdown: dict[str, dict]   # {month: {category: amount}}
    insights: list[PatternInsight]


# ── Cell 4: Validation & cleaning ────────────────────────────

def _validate(df: pd.DataFrame) -> None:
    missing = REQUIRED_COLUMNS - set(df.columns)
    if missing:
        raise ValueError(f"Missing required column(s): {', '.join(sorted(missing))}")
    if df.empty:
        raise ValueError("DataFrame is empty — nothing to analyse.")


def _clean(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()

    # Normalise Amount
    df["Amount"] = (
        df["Amount"].astype(str)
        .str.replace(r"[$,]", "", regex=True)
        .str.strip()
    )
    df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")

    # Parse Date if present
    if "Date" in df.columns:
        df["Date"] = pd.to_datetime(df["Date"], errors="coerce")
        df["YearMonth"] = df["Date"].dt.to_period("M").astype(str)

    # Normalise Category
    df["Category"] = df["Category"].astype(str).str.strip().str.title()

    before = len(df)
    df = df.dropna(subset=["Amount", "Category"])
    df = df[df["Amount"] > 0]   # keep charges only
    dropped = before - len(df)
    if dropped:
        print(f"⚠️  Dropped {dropped} row(s) with missing/zero/negative amounts.")

    return df


# ── Cell 5: Core analysis ─────────────────────────────────────

def _build_insights(
    category_percentages: pd.Series,
    category_totals: pd.Series,
    outliers: list[str],
    low: list[str],
    high: list[str],
) -> list[PatternInsight]:
    insights: list[PatternInsight] = []
    top_cat = category_percentages.idxmax()
    top_pct = category_percentages.max()

    # Concentration
    if top_pct > THRESHOLDS["high_concentration"]:
        insights.append(PatternInsight(
            level="warning",
            title=f"{top_cat} dominates at {top_pct:.1f}% of spend",
            detail="Heavy concentration in one category makes budgets fragile. "
                   "Review whether this reflects a strategic priority or runaway cost.",
        ))
    elif top_pct < THRESHOLDS["well_distributed"]:
        insights.append(PatternInsight(
            level="success",
            title="Spending is well distributed",
            detail=f"No single category exceeds {top_pct:.1f}%, "
                   "reducing dependency risk and giving you flexible room to adjust.",
        ))
    else:
        insights.append(PatternInsight(
            level="info",
            title=f"Top category '{top_cat}' holds {top_pct:.1f}% of spend",
            detail="This is within a normal range, but worth monitoring for upward drift.",
        ))

    # Outlier categories
    if outliers:
        insights.append(PatternInsight(
            level="warning",
            title=f"Outlier spend detected: {', '.join(outliers)}",
            detail="These categories are more than 2× the average category spend. "
                   "Investigate whether this is intentional or driven by one-off events.",
        ))

    # Low-spend categories
    if low:
        insights.append(PatternInsight(
            level="info",
            title=f"Low allocation (<{THRESHOLDS['low_spend']:.0f}%): {', '.join(low)}",
            detail="These may be underfunded or simply low-priority. "
                   "Confirm they're receiving the attention they need.",
        ))

    # High-spend categories (top 3 excluding the overall top)
    if high:
        insights.append(PatternInsight(
            level="info",
            title=f"Other high-spend areas: {', '.join(high)}",
            detail="These categories account for a significant share after the top category. "
                   "Bundling savings efforts across these could compound quickly.",
        ))

    return insights


def analyze_spending_patterns(df: pd.DataFrame) -> SpendingPatternResult:
    """
    Analyse spending patterns from a cleaned transaction DataFrame.

    Parameters
    ----------
    df : DataFrame with at least 'Category' and 'Amount' columns.
         'Date' column is optional but enables monthly breakdown.

    Returns
    -------
    SpendingPatternResult dataclass with all metrics and insights.
    """
    _validate(df)
    df = _clean(df)

    # ── Aggregations ─────────────────────────────────────────
    category_totals = (
        df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
    )
    total_spending      = category_totals.sum()
    category_percentages = (category_totals / total_spending * 100).round(2)
    category_tx_counts  = df.groupby("Category")["Amount"].count().to_dict()
    category_avg_tx     = (
        df.groupby("Category")["Amount"].mean().round(2).to_dict()
    )

    # ── Outlier detection ─────────────────────────────────────
    mean_spend = category_totals.mean()
    outliers   = category_totals[
        category_totals > mean_spend * THRESHOLDS["spike_vs_mean"]
    ].index.tolist()

    # ── Threshold buckets ─────────────────────────────────────
    low_spend  = category_percentages[
        category_percentages < THRESHOLDS["low_spend"]
    ].index.tolist()
    high_spend = category_percentages[
        (category_percentages >= THRESHOLDS["well_distributed"]) &
        (category_percentages.index != category_totals.idxmax())
    ].index.tolist()

    # ── Monthly breakdown (if Date column present) ────────────
    monthly_breakdown: dict[str, dict] = {}
    if "YearMonth" in df.columns:
        mb = (
            df.groupby(["YearMonth", "Category"])["Amount"]
            .sum().unstack(fill_value=0)
        )
        monthly_breakdown = {
            month: row.to_dict() for month, row in mb.iterrows()
        }

    # ── Insights ──────────────────────────────────────────────
    insights = _build_insights(
        category_percentages, category_totals, outliers, low_spend, high_spend
    )

    top_cat = category_totals.idxmax()
    return SpendingPatternResult(
        total_spending=round(total_spending, 2),
        category_totals=category_totals.round(2).to_dict(),
        category_percentages=category_percentages.to_dict(),
        category_tx_counts=category_tx_counts,
        category_avg_tx=category_avg_tx,
        top_category={
            "name":       top_cat,
            "amount":     round(category_totals[top_cat], 2),
            "percentage": round(category_percentages[top_cat], 2),
        },
        outlier_categories=outliers,
        low_spend_categories=low_spend,
        high_spend_categories=high_spend,
        monthly_breakdown=monthly_breakdown,
        insights=insights,
    )


# ── Cell 6: HTML renderer ─────────────────────────────────────

_LEVEL_STYLE = {
    "warning": ("⚠️",  "#fef2f2", "#dc2626", "#fee2e2"),
    "success": ("✅",  "#f0fdf4", "#16a34a", "#dcfce7"),
    "info":    ("ℹ️",  "#eff6ff", "#2563eb", "#dbeafe"),
}

def _bar(pct: float, color: str = "#4f46e5", max_pct: float = 100) -> str:
    width = min(pct / max_pct * 100, 100)
    return (
        f'<div style="background:#e5e7eb;border-radius:4px;height:10px;width:100%">'
        f'<div style="background:{color};width:{width:.1f}%;height:10px;border-radius:4px"></div>'
        f'</div>'
    )

def _fmt(v: float, currency: str = "AUD") -> str:
    return f"{currency} ${v:,.2f}"


def render_pattern_report(
    result: SpendingPatternResult,
    currency: str = "AUD",
) -> None:
    """Display a rich HTML report in Colab."""

    # ── KPI strip ────────────────────────────────────────────
    n_cats = len(result.category_totals)
    avg_cat = result.total_spending / n_cats if n_cats else 0
    kpis = [
        ("Total Spending",    _fmt(result.total_spending, currency),  "#10b981"),
        ("Categories",        str(n_cats),                            "#4f46e5"),
        ("Top Category",      result.top_category["name"],            "#f59e0b"),
        ("Top Category %",    f"{result.top_category['percentage']}%","#ef4444"),
        ("Avg per Category",  _fmt(avg_cat, currency),                "#6366f1"),
    ]
    cards = "".join(f"""
    <div style="background:#fff;border-radius:10px;padding:14px 18px;
                box-shadow:0 1px 4px rgba(0,0,0,.08);text-align:center;min-width:130px">
      <div style="font-size:11px;color:#6b7280;margin-bottom:4px">{l}</div>
      <div style="font-size:18px;font-weight:700;color:{c}">{v}</div>
    </div>""" for l, v, c in kpis)

    display(HTML(f"""
    <h2 style="font-family:sans-serif;margin-bottom:8px">📊 Spending Pattern Analysis</h2>
    <div style="display:flex;flex-wrap:wrap;gap:10px;margin-bottom:20px">{cards}</div>
    """))

    # ── Category breakdown table ──────────────────────────────
    max_val = max(result.category_totals.values())
    rows = ""
    colors = ["#4f46e5","#7c3aed","#2563eb","#0891b2","#059669","#d97706","#dc2626"]
    for i, (cat, amt) in enumerate(result.category_totals.items()):
        pct   = result.category_percentages[cat]
        count = result.category_tx_counts.get(cat, 0)
        avg   = result.category_avg_tx.get(cat, 0)
        color = colors[i % len(colors)]
        rows += f"""<tr style="border-bottom:1px solid #f3f4f6">
          <td style="padding:9px 12px;font-family:sans-serif;font-size:14px">
            <span style="display:inline-block;width:10px;height:10px;border-radius:50%;
                         background:{color};margin-right:6px"></span>{cat}
          </td>
          <td style="padding:9px 12px;text-align:right;font-weight:600;font-size:14px">
            {_fmt(amt, currency)}</td>
          <td style="padding:9px 12px;width:180px">{_bar(amt, color, max_val)}</td>
          <td style="padding:9px 12px;color:#6b7280;font-size:13px">{pct:.1f}%</td>
          <td style="padding:9px 12px;color:#6b7280;font-size:13px">{count} txns</td>
          <td style="padding:9px 12px;color:#6b7280;font-size:13px">
            avg {_fmt(avg, currency)}</td>
        </tr>"""

    display(HTML(f"""
    <h3 style="font-family:sans-serif">📂 Category Breakdown</h3>
    <table style="border-collapse:collapse;width:100%">
      <thead>
        <tr style="background:#f9fafb;font-family:sans-serif;font-size:12px;color:#6b7280">
          <th style="padding:8px 12px;text-align:left">Category</th>
          <th style="padding:8px 12px;text-align:right">Total</th>
          <th style="padding:8px 12px">Share</th>
          <th style="padding:8px 12px">%</th>
          <th style="padding:8px 12px">Txns</th>
          <th style="padding:8px 12px">Avg Txn</th>
        </tr>
      </thead>
      <tbody>{rows}</tbody>
    </table>
    """))

    # ── Monthly breakdown ─────────────────────────────────────
    if result.monthly_breakdown:
        months = sorted(result.monthly_breakdown)
        cats   = list(result.category_totals.keys())
        header = "".join(f'<th style="padding:6px 10px;font-size:12px">{c}</th>' for c in cats)
        m_rows = ""
        for month in months:
            data = result.monthly_breakdown[month]
            cells = "".join(
                f'<td style="padding:6px 10px;text-align:right;font-size:13px">'
                f'{_fmt(data.get(cat, 0), currency)}</td>'
                for cat in cats
            )
            m_rows += f'<tr style="border-bottom:1px solid #f3f4f6"><td style="padding:6px 10px;font-family:monospace;font-size:13px">{month}</td>{cells}</tr>'

        display(HTML(f"""
        <h3 style="font-family:sans-serif;margin-top:20px">📅 Monthly Category Breakdown</h3>
        <div style="overflow-x:auto">
        <table style="border-collapse:collapse;width:100%">
          <thead>
            <tr style="background:#f9fafb;color:#6b7280">
              <th style="padding:6px 10px;text-align:left;font-size:12px">Month</th>{header}
            </tr>
          </thead>
          <tbody>{m_rows}</tbody>
        </table>
        </div>
        """))

    # ── Insight cards ─────────────────────────────────────────
    display(HTML("<h3 style='font-family:sans-serif;margin-top:20px'>💡 Insights</h3>"))
    for ins in result.insights:
        icon, bg, accent, badge_bg = _LEVEL_STYLE[ins.level]
        display(HTML(f"""
        <div style="background:{bg};border-left:4px solid {accent};border-radius:8px;
                    padding:12px 16px;margin-bottom:10px;font-family:sans-serif">
          <div style="display:flex;align-items:center;gap:8px;margin-bottom:4px">
            <span style="background:{badge_bg};color:{accent};font-size:11px;font-weight:700;
                         padding:2px 8px;border-radius:12px">{icon} {ins.level.upper()}</span>
          </div>
          <div style="font-weight:700;font-size:14px;margin-bottom:4px">{ins.title}</div>
          <div style="font-size:13px;color:#374151">{ins.detail}</div>
        </div>
        """))


# ── Cell 7: Upload widget + runner ────────────────────────────

def run_pattern_analysis(
    source=None,
    currency: str = "AUD",
) -> SpendingPatternResult | None:
    """
    Analyse spending patterns from a CSV.

    Parameters
    ----------
    source   : file path (str), bytes, file-like object, or None for upload widget.
    currency : currency label for display (default "AUD").

    Returns
    -------
    SpendingPatternResult, or None on failure.
    """

    def _run(src) -> SpendingPatternResult | None:
        try:
            if isinstance(src, bytes):
                df = pd.read_csv(io.BytesIO(src))
            elif isinstance(src, str):
                df = pd.read_csv(src)
            else:
                df = pd.read_csv(src)

            result = analyze_spending_patterns(df)
            render_pattern_report(result, currency=currency)
            display(HTML("<h3 style='font-family:sans-serif;margin-top:20px'>🔍 Raw Data Preview</h3>"))
            display(df.head(10))
            return result

        except (ValueError, FileNotFoundError) as e:
            display(HTML(f"<p style='color:#dc2626;font-family:sans-serif'>❌ {e}</p>"))
        except Exception as e:
            display(HTML(f"<p style='color:#dc2626;font-family:sans-serif'>❌ Unexpected error: {e}</p>"))
        return None

    # ── Upload widget ─────────────────────────────────────────
    if source is None:
        uploader = widgets.FileUpload(accept=".csv", multiple=False,
                                      description="Upload CSV")
        out = widgets.Output()

        def on_upload(change):
            with out:
                clear_output(wait=True)
                file_bytes = list(uploader.value.values())[0]["content"]
                _run(bytes(file_bytes))

        uploader.observe(on_upload, names="value")
        display(uploader, out)
        return None

    return _run(source)


# ── Cell 8: Run ───────────────────────────────────────────────
# Option A — interactive upload (shows file-picker button):
run_pattern_analysis()

# Option B — load from Google Drive:
# run_pattern_analysis("/content/drive/MyDrive/spending.csv")

# Option C — pass a DataFrame directly:
# result = analyze_spending_patterns(my_df)
# render_pattern_report(result, currency="USD")

# Option D — sample data to test without a CSV:
# _sample_df = pd.DataFrame({
#     "Date":        ["2025-03-01","2025-03-15","2025-04-02","2025-04-20","2025-04-25"],
#     "Category":    ["Food","Transport","Food","Entertainment","Utilities"],
#     "Amount":      [1200, 350, 980, 520, 180],
#     "Description": ["Groceries","Bus pass","Restaurants","Cinema","Electricity"],
# })
# run_pattern_analysis(_sample_df)

# Test your function
# analysis = analyze_spending_patterns(clean_data)
# print(analysis)
````

````python
# 🤖 AI Collaboration: Business Insights Generator
# Ask AI to help create professional financial recommendations

def generate_financial_recommendations(analysis_data):
    """
    Generate actionable financial recommendations based on spending analysis

    🤖 AI Collaboration Prompt:
    "Based on spending analysis data, create professional financial
    recommendations. Include specific savings opportunities, spending
    pattern observations, and actionable advice formatted for a
    personal finance app user."

    Args:
        analysis_data: Dictionary with spending analysis results
    Returns:
        str: Formatted recommendations report
    """
    # Your AI-assisted implementation goes here
    # Focus on actionable, user-friendly advice
    # ============================================================
# 💡 Smart Finance Recommendations — Google Colab Edition
# Paste after finance_analyzer_colab.py cells, then run.
# ============================================================

# ── Cell 8: Imports ─────────────────────────────────────────
import re
from dataclasses import dataclass

# ipywidgets, IPython.display, pandas already imported from analyzer cells

# ── Cell 9: Thresholds & tips ────────────────────────────────

THRESHOLDS = {
    "high_single_category_pct": 40.0,  # warn if one category > 40% of spend
    "high_refund_ratio":         0.10,  # flag if refunds > 10% of gross
    "low_transaction_count":     5,     # flag suspiciously few transactions
    "high_avg_transaction":    500.0,   # flag high average (currency units)
    "monthly_spike_pct":        15.0,   # flag month-on-month rise > 15%
}

SAVINGS_TIPS = {
    "Food":          "Meal-plan weekly and batch-cook to cut impulse takeaway orders.",
    "Groceries":     "Use a shopping list and compare unit prices to avoid overspending.",
    "Transport":     "Consider a monthly transit pass or carpooling to reduce per-trip costs.",
    "Entertainment": "Audit streaming subscriptions — cancel any unused in the last 30 days.",
    "Shopping":      "Apply a 48-hour rule before non-essential purchases to curb impulse buys.",
    "Travel":        "Book 6–8 weeks ahead and set fare alerts to lock in lower prices.",
    "Health":        "Check insurer coverage before paying out-of-pocket for services.",
    "Utilities":     "Compare providers annually — small tariff savings compound over time.",
}


# ── Cell 10: Data types ──────────────────────────────────────

@dataclass
class Recommendation:
    priority: str   # "HIGH" | "MEDIUM" | "LOW"
    category: str   # e.g. "Budget Alert"
    title: str
    detail: str
    action: str


# ── Cell 11: Helpers ─────────────────────────────────────────

def _parse_amount(value) -> float:
    """Coerce pre-formatted strings ('AUD $1,234.00') or raw floats to float."""
    if isinstance(value, (int, float)):
        return float(value)
    cleaned = re.sub(r"[^\d.\-]", "", str(value))
    try:
        return float(cleaned)
    except ValueError:
        return 0.0


def _fmt(amount: float, currency: str) -> str:
    return f"{currency} ${amount:,.2f}"


# ── Cell 12: Recommendation engine ──────────────────────────

def build_recommendations(data: dict) -> list[Recommendation]:
    """
    Inspect analysis_data and return a prioritised list of Recommendation objects.
    Works with both the old string-formatted dict and the improved numeric dict.
    """
    recs: list[Recommendation] = []

    # Support both old keys ("Total Spending") and new keys ("total_spending")
    total_spending    = _parse_amount(data.get("total_spending")    or data.get("Total Spending",    0))
    total_refunds     = abs(_parse_amount(data.get("total_refunds") or data.get("Refund Impact",     0)))
    avg_transaction   = _parse_amount(data.get("avg_transaction")   or data.get("Average Transaction", 0))
    transaction_count = int(data.get("transaction_count")           or data.get("Transaction Count", 0))
    top_categories    = data.get("top_categories")                  or data.get("Top Categories",   {})
    pct_by_category   = data.get("pct_by_category", {})
    monthly_totals    = data.get("monthly_totals",  {})
    currency          = data.get("currency", "AUD")

    # ── 1. High single-category concentration ───────────────
    for cat, val in top_categories.items():
        pct = pct_by_category.get(cat, 0)
        if pct >= THRESHOLDS["high_single_category_pct"]:
            tip = SAVINGS_TIPS.get(cat, "Review this category for discretionary cuts.")
            recs.append(Recommendation(
                priority="HIGH",
                category="Budget Alert",
                title=f"{cat} is {pct:.0f}% of your total spend",
                detail=(f"You spent {_fmt(val, currency)} on {cat} — your single largest outlay. "
                        "Heavy concentration in one area makes budgets fragile."),
                action=tip,
            ))

    # ── 2. Category-specific tips (non-dominant categories) ─
    for cat, val in top_categories.items():
        pct = pct_by_category.get(cat, 0)
        if cat in SAVINGS_TIPS and pct < THRESHOLDS["high_single_category_pct"]:
            recs.append(Recommendation(
                priority="MEDIUM",
                category="Category Insight",
                title=f"Optimise your {cat} spend ({pct:.0f}%)",
                detail=f"You spent {_fmt(val, currency)} on {cat} this period.",
                action=SAVINGS_TIPS[cat],
            ))

    # ── 3. High average transaction ─────────────────────────
    if avg_transaction > THRESHOLDS["high_avg_transaction"]:
        recs.append(Recommendation(
            priority="MEDIUM",
            category="Transaction Pattern",
            title=f"High average transaction: {_fmt(avg_transaction, currency)}",
            detail=("Your average transaction is above typical day-to-day spend, "
                    "suggesting large or irregular purchases are skewing your budget."),
            action=("Plan major purchases in advance and allocate a dedicated monthly savings "
                    "buffer so they don't disrupt cash flow."),
        ))

    # ── 4. High refund ratio ─────────────────────────────────
    if total_spending > 0:
        refund_ratio = total_refunds / total_spending
        if refund_ratio >= THRESHOLDS["high_refund_ratio"]:
            recs.append(Recommendation(
                priority="MEDIUM",
                category="Refund Tracking",
                title=f"Refunds are {refund_ratio*100:.0f}% of gross spend",
                detail=(f"You received {_fmt(total_refunds, currency)} back. "
                        "Frequent returns can signal impulse buying or product dissatisfaction."),
                action=("Note what you're returning. If it's the same category repeatedly, "
                        "reconsider how you shop there before buying."),
            ))

    # ── 5. Monthly spending trend ────────────────────────────
    if len(monthly_totals) >= 2:
        months   = sorted(monthly_totals)
        recent   = monthly_totals[months[-1]]
        previous = monthly_totals[months[-2]]
        change   = (recent - previous) / previous * 100 if previous else 0
        if change > THRESHOLDS["monthly_spike_pct"]:
            recs.append(Recommendation(
                priority="HIGH",
                category="Spending Trend",
                title=f"Spending jumped {change:.0f}% vs last month",
                detail=(f"{months[-2]}: {_fmt(previous, currency)} → "
                        f"{months[-1]}: {_fmt(recent, currency)}. "
                        "A sharp rise is worth investigating before it becomes a habit."),
                action=("Identify whether the jump is a one-off large purchase or a new recurring "
                        "expense. Set a monthly cap alert in your banking app."),
            ))
        elif change < -THRESHOLDS["monthly_spike_pct"]:
            recs.append(Recommendation(
                priority="LOW",
                category="Spending Trend",
                title=f"Spending dropped {abs(change):.0f}% vs last month 🎉",
                detail=(f"{months[-2]}: {_fmt(previous, currency)} → "
                        f"{months[-1]}: {_fmt(recent, currency)}."),
                action=("Great discipline! Redirect the difference into an emergency fund "
                        "or high-interest savings account."),
            ))

    # ── 6. Low transaction count ─────────────────────────────
    if 0 < transaction_count <= THRESHOLDS["low_transaction_count"]:
        recs.append(Recommendation(
            priority="LOW",
            category="Data Quality",
            title="Few transactions analysed",
            detail=(f"Only {transaction_count} transaction(s) were found. "
                    "Insights may not reflect your full spending picture."),
            action=("Ensure your CSV covers a full month and includes all payment methods: "
                    "card, cash, and digital wallets."),
        ))

    # ── 7. Universal best-practice tips ─────────────────────
    recs.append(Recommendation(
        priority="LOW",
        category="Best Practice",
        title="Build a 3-month emergency fund",
        detail="Financial resilience starts with a buffer equal to 3 months of essential expenses.",
        action=f"Target saving {_fmt(total_spending * 3, currency)} as your emergency baseline.",
    ))
    recs.append(Recommendation(
        priority="LOW",
        category="Best Practice",
        title="Review subscriptions every quarter",
        detail="Recurring charges accumulate quietly and are easy to forget.",
        action="List every recurring charge this month and cancel anything unused in the last 30 days.",
    ))

    # Sort HIGH → MEDIUM → LOW
    order = {"HIGH": 0, "MEDIUM": 1, "LOW": 2}
    recs.sort(key=lambda r: order[r.priority])
    return recs


# ── Cell 13: HTML renderer ───────────────────────────────────

_PRIORITY_STYLE = {
    "HIGH":   ("🔴", "#fef2f2", "#dc2626", "#fee2e2"),
    "MEDIUM": ("🟡", "#fffbeb", "#d97706", "#fef3c7"),
    "LOW":    ("🟢", "#f0fdf4", "#16a34a", "#dcfce7"),
}


def render_recommendations(recs: list[Recommendation], data: dict) -> None:
    currency = data.get("currency", "AUD")
    total    = _parse_amount(data.get("total_spending") or data.get("Total Spending", 0))

    display(HTML("""
    <h2 style='font-family:sans-serif;margin-bottom:4px'>💡 Financial Recommendations</h2>
    <p style='font-family:sans-serif;color:#6b7280;margin-top:0;font-size:13px'>
      Personalised to your spending · sorted by priority
    </p>
    """))

    for rec in recs:
        icon, bg, accent, badge_bg = _PRIORITY_STYLE[rec.priority]
        display(HTML(f"""
        <div style="background:{bg};border-left:4px solid {accent};border-radius:8px;
                    padding:14px 18px;margin-bottom:12px;font-family:sans-serif">
          <div style="display:flex;align-items:center;gap:8px;margin-bottom:6px">
            <span style="background:{badge_bg};color:{accent};font-size:11px;font-weight:700;
                         padding:2px 8px;border-radius:12px">{icon} {rec.priority}</span>
            <span style="font-size:11px;color:#6b7280">{rec.category}</span>
          </div>
          <div style="font-weight:700;font-size:15px;margin-bottom:4px">{rec.title}</div>
          <div style="font-size:13px;color:#374151;margin-bottom:8px">{rec.detail}</div>
          <div style="background:rgba(255,255,255,.7);border-radius:6px;padding:8px 12px;
                      font-size:13px;color:{accent}">
            <strong>Action:</strong> {rec.action}
          </div>
        </div>
        """))

    high_count   = sum(1 for r in recs if r.priority == "HIGH")
    medium_count = sum(1 for r in recs if r.priority == "MEDIUM")

    display(HTML(f"""
    <div style="margin-top:16px;font-family:sans-serif;font-size:12px;color:#9ca3af;
                border-top:1px solid #e5e7eb;padding-top:10px">
      {len(recs)} recommendation(s) &nbsp;·&nbsp;
      🔴 {high_count} high &nbsp;·&nbsp;
      🟡 {medium_count} medium &nbsp;·&nbsp;
      based on {_fmt(total, currency)} analysed
    </div>
    """))


# ── Cell 14: Public entry point ──────────────────────────────

def generate_financial_recommendations(
    analysis_data: dict,
    *,
    render: bool = True,
) -> list[dict]:
    """
    Generate and display prioritised recommendations from spending analysis output.

    Parameters
    ----------
    analysis_data : dict returned by analyze_spending_data()
    render        : if True (default), displays rich HTML cards in Colab

    Returns
    -------
    List of recommendation dicts for further processing or export
    """
    recs = build_recommendations(analysis_data)

    if render:
        render_recommendations(recs, analysis_data)

    return [
        {
            "priority": r.priority,
            "category": r.category,
            "title":    r.title,
            "detail":   r.detail,
            "action":   r.action,
        }
        for r in recs
    ]


# ── Cell 15: Run ─────────────────────────────────────────────
# After running the analyzer cells, simply do:
#
#   result = analyze_spending_data()
#   generate_financial_recommendations(result)
#
# Or test standalone with sample data:

_sample = {
    "currency":         "AUD",
    "total_spending":   3200.00,
    "total_refunds":    -420.00,
    "net_spending":     2780.00,
    "avg_transaction":   620.00,
    "transaction_count": 18,
    "spending_count":    15,
    "refund_count":       3,
    "top_categories":   {"Food": 1400.00, "Transport": 600.00, "Entertainment": 520.00},
    "pct_by_category":  {"Food": 43.8, "Transport": 18.8, "Entertainment": 16.3},
    "monthly_totals":   {"2025-03": 2600.00, "2025-04": 3200.00},
    "largest_transaction": {
        "description": "Annual gym membership",
        "amount":      900.00,
        "category":    "Health",
        "date":        "2025-04-03",
    },
}

generate_financial_recommendations(_sample)

# Test your function
# recommendations = generate_financial_recommendations(analysis)
# print(recommendations)
````

````python
# 🤖 AI Collaboration: Financial Advice Chatbot
# Ask AI to help you create a finance-focused chat personality

from hands_on_ai import chat

def create_finance_chat_personality():
    """
    Set up a financial advisor chat personality using hands-on-ai

    🤖 AI Collaboration Prompt:
    "Help me create a system prompt for a friendly, professional financial
    advisor chatbot that can provide spending advice based on transaction
    analysis. The personality should be encouraging but practical."
    """
 # ============================================================
# 🤖 Smart Finance Chat — Google Colab Edition
# Paste after previous analyzer + recommendations cells.
# ============================================================

# ── Cell 16: Install & import ────────────────────────────────
# !pip install hands-on-ai --quiet

from hands_on_ai.chat import get_response   # ✅ correct API — no chat.setup()
import ipywidgets as widgets
from IPython.display import display, HTML, clear_output
import datetime, re
from dataclasses import dataclass, field


# ── Cell 17: System prompt builder ───────────────────────────

def build_system_prompt(analysis_data: dict | None = None) -> str:
    """
    Build a rich system prompt, optionally injecting real spending context
    so the AI gives data-aware answers rather than generic advice.
    """
    base = """
You are Morgan, a friendly and professional personal finance advisor.

## Personality
- Warm, encouraging, and non-judgmental — money is stressful; make it manageable
- Practical: every suggestion should be actionable today or this week
- Honest: if spending looks risky, say so clearly but constructively
- Concise: prefer short paragraphs or a tight bullet list over walls of text

## Expertise
- Personal budgeting and cash-flow management
- Spending pattern analysis and category optimisation
- Savings strategies (emergency funds, sinking funds, automation)
- Debt reduction (avalanche vs snowball methods)
- Subscription and recurring-charge audits
- Refund tracking and its impact on net spend

## Response rules
- Always end with one clear, specific next step the user can take
- Never recommend specific financial products or investment vehicles
- If asked something outside personal finance, politely redirect
- Use everyday language; define any financial term you use
- Keep responses under 200 words unless the user asks for detail
""".strip()

    if not analysis_data:
        return base

    # ── Inject real spending numbers ──────────────────────────
    currency        = analysis_data.get("currency", "AUD")
    total_spending  = analysis_data.get("total_spending", 0)
    net_spending    = analysis_data.get("net_spending", 0)
    avg_transaction = analysis_data.get("avg_transaction", 0)
    tx_count        = analysis_data.get("transaction_count", 0)
    top_cats        = analysis_data.get("top_categories", {})
    pct             = analysis_data.get("pct_by_category", {})
    monthly         = analysis_data.get("monthly_totals", {})
    largest         = analysis_data.get("largest_transaction", {})

    cats_text = "\n".join(
        f"  - {cat}: {currency} ${val:,.2f} ({pct.get(cat, 0):.1f}%)"
        for cat, val in top_cats.items()
    ) or "  - No category data"

    monthly_text = "\n".join(
        f"  - {m}: {currency} ${v:,.2f}" for m, v in sorted(monthly.items())
    ) or "  - No monthly data"

    largest_text = (
        f"  {largest.get('description','N/A')} — "
        f"{currency} ${largest.get('amount', 0):,.2f} "
        f"({largest.get('category','')}, {largest.get('date','')})"
        if largest else "  None recorded"
    )

    context = f"""

## User's Spending Data (from uploaded CSV)
Use this to give personalised answers. Refer to it naturally ("I can see your top category is…")
rather than listing raw figures. Focus on what the numbers mean for their financial health.

- Gross spend     : {currency} ${total_spending:,.2f}
- Net spend       : {currency} ${net_spending:,.2f}
- Avg transaction : {currency} ${avg_transaction:,.2f}
- Transactions    : {tx_count}

Top categories:
{cats_text}

Monthly breakdown:
{monthly_text}

Largest transaction:
{largest_text}
"""
    return base + context


# ── Cell 18: Session class ────────────────────────────────────

@dataclass
class FinanceChatSession:
    """
    Wraps hands_on_ai get_response() with history, a system prompt,
    and an optional Colab HTML display.
    """

    STARTERS = [
        "Where am I overspending?",
        "How do I build an emergency fund?",
        "Which category should I cut first?",
        "How do my refunds affect my budget?",
        "Tips for reducing my biggest expense?",
        "How can I save more each month?",
    ]

    analysis_data: dict | None = None
    history: list[dict] = field(default_factory=list)
    _system_prompt: str = field(init=False)

    def __post_init__(self):
        self._system_prompt = build_system_prompt(self.analysis_data)

    # ── Helpers ───────────────────────────────────────────────

    def _ts(self) -> str:
        return datetime.datetime.now().strftime("%H:%M")

    def _bubble(self, role: str, text: str, time: str) -> str:
        is_user   = role == "user"
        align     = "flex-end"  if is_user else "flex-start"
        bg        = "#4f46e5"   if is_user else "#f3f4f6"
        color     = "#ffffff"   if is_user else "#111827"
        label     = "You"       if is_user else "Morgan 🤖"
        safe_text = text.replace("\n", "<br>")
        radius    = "16px 16px 4px 16px" if is_user else "16px 16px 16px 4px"
        side      = "right" if is_user else "left"
        return f"""
        <div style="display:flex;justify-content:{align};margin-bottom:10px">
          <div style="max-width:75%">
            <div style="font-size:11px;color:#9ca3af;margin-bottom:2px;text-align:{side}">
              {label} · {time}
            </div>
            <div style="background:{bg};color:{color};padding:10px 14px;
                        border-radius:{radius};font-family:sans-serif;
                        font-size:14px;line-height:1.5">
              {safe_text}
            </div>
          </div>
        </div>"""

    def _render(self, output_area) -> None:
        with output_area:
            clear_output(wait=True)
            if not self.history:
                display(HTML("""
                <div style="font-family:sans-serif;color:#9ca3af;text-align:center;
                            padding:40px;font-size:14px">
                  Ask Morgan anything about your finances 👆
                </div>"""))
                return
            html = "".join(self._bubble(m["role"], m["text"], m["time"])
                           for m in self.history)
            display(HTML(f'<div style="padding:8px">{html}</div>'))

    # ── Public API ────────────────────────────────────────────

    def send(self, message: str, output_area=None) -> str:
        """Send a message and get a response from hands_on_ai."""
        message = message.strip()
        if not message:
            return ""

        self.history.append({"role": "user", "text": message, "time": self._ts()})
        if output_area:
            self._render(output_area)

        try:
            # ✅ Correct hands_on_ai API:
            #    get_response(prompt, system=system_prompt)
            response = get_response(message, system=self._system_prompt)
        except TypeError:
            # Older versions may not accept `system` as keyword arg
            try:
                response = get_response(message, self._system_prompt)
            except Exception as e:
                response = f"⚠️ Could not reach the AI. ({e})"
        except Exception as e:
            response = f"⚠️ Sorry, something went wrong. ({e})"

        self.history.append({"role": "assistant", "text": response, "time": self._ts()})
        if output_area:
            self._render(output_area)
        return response

    def clear(self) -> None:
        """Reset conversation history (system prompt is preserved)."""
        self.history.clear()


# ── Cell 19: Colab UI ─────────────────────────────────────────

def launch_finance_chat(analysis_data: dict | None = None) -> FinanceChatSession:
    """
    Render a full chat UI in Colab.

    Parameters
    ----------
    analysis_data : dict returned by analyze_spending_data() — optional but
                    strongly recommended for personalised answers.

    Returns
    -------
    FinanceChatSession (can also call session.send() programmatically).
    """
    session = FinanceChatSession(analysis_data=analysis_data)

    has_data    = bool(analysis_data)
    badge       = ("✅ Spending data loaded — Morgan knows your numbers"
                   if has_data else
                   "ℹ️ No spending data — Morgan will give general advice")
    badge_color = "#10b981" if has_data else "#f59e0b"

    # ── Header ───────────────────────────────────────────────
    display(HTML(f"""
    <div style="font-family:sans-serif;border:1px solid #e5e7eb;border-radius:12px;
                overflow:hidden;margin-bottom:8px">
      <div style="background:#4f46e5;padding:14px 18px;color:#fff;
                  display:flex;align-items:center;gap:10px">
        <span style="font-size:22px">🤖</span>
        <div>
          <div style="font-weight:700;font-size:16px">Morgan — Finance Advisor</div>
          <div style="font-size:12px;opacity:.8">Powered by hands-on-ai</div>
        </div>
      </div>
      <div style="background:#f9fafb;padding:8px 16px;font-size:12px;color:{badge_color}">
        {badge}
      </div>
    </div>
    """))

    # ── Conversation area ────────────────────────────────────
    output_area = widgets.Output(
        layout=widgets.Layout(
            height="340px", overflow_y="auto",
            border="1px solid #e5e7eb", border_radius="8px", padding="4px",
        )
    )
    session._render(output_area)
    display(output_area)

    # ── Starter chips ────────────────────────────────────────
    chips = [
        widgets.Button(
            description=q,
            layout=widgets.Layout(width="auto", height="30px"),
            style=widgets.ButtonStyle(button_color="#ede9fe", font_color="#4f46e5"),
        )
        for q in FinanceChatSession.STARTERS
    ]
    display(widgets.HBox(chips, layout=widgets.Layout(flex_wrap="wrap", gap="6px", margin="8px 0")))

    # ── Input row ────────────────────────────────────────────
    text_input = widgets.Text(
        placeholder="Ask Morgan about your finances…",
        layout=widgets.Layout(flex="1", height="36px"),
    )
    send_btn  = widgets.Button(description="Send",  button_style="primary",
                               layout=widgets.Layout(width="80px", height="36px"))
    clear_btn = widgets.Button(description="Clear",
                               layout=widgets.Layout(width="70px", height="36px"),
                               style=widgets.ButtonStyle(button_color="#f3f4f6"))
    display(widgets.HBox([text_input, send_btn, clear_btn],
                         layout=widgets.Layout(gap="6px", margin="4px 0")))

    # ── Handlers ──────────────────────────────────────────────
    def on_send(_=None):
        msg = text_input.value.strip()
        if not msg:
            return
        text_input.value = ""
        send_btn.disabled = clear_btn.disabled = True
        session.send(msg, output_area)
        send_btn.disabled = clear_btn.disabled = False

    def on_clear(_):
        session.clear()
        session._render(output_area)

    def make_chip_handler(q):
        def handler(_):
            text_input.value = q
            on_send()
        return handler

    send_btn.on_click(on_send)
    text_input.on_submit(on_send)
    clear_btn.on_click(on_clear)
    for chip in chips:
        chip.on_click(make_chip_handler(chip.description))

    return session


# ── Cell 20: Run ─────────────────────────────────────────────
# With spending data (personalised):
#   result  = analyze_spending_data()
#   session = launch_finance_chat(result)
#
# Without spending data (general advice):
#   session = launch_finance_chat()
#
# Programmatic (no UI):
#   session = FinanceChatSession(analysis_data=result)
#   reply   = session.send("Where am I overspending?")
#   print(reply)

session = launch_finance_chat()   # swap in `result` once analyzer has run

# Test your chatbot
# response = chat.say("I spend too much on coffee, what should I do?")
# print(response)
````

````python
# 🤖 AI Collaboration: Document Retrieval Setup
# Ask AI to help you set up RAG for financial documents
# ============================================================
# CELL 1 — Install
# ============================================================




# ============================================================
# CELL 3 — Create sample financial documents
# ============================================================
DOCS_DIR = Path("financial_docs")
DOCS_DIR.mkdir(exist_ok=True)

docs = {
    "q1_2024_report.txt": (
        "Q1 2024 Financial Report\n"
        "Total Revenue: $45,200\n"
        "Total Expenses: $32,800\n"
        "Net Savings: $12,400\n"
        "Top expense categories: Housing (38%), Food (18%), Transport (12%)\n"
    ),
    "budgeting_50_30_20.txt": (
        "The 50/30/20 Budgeting Rule\n"
        "50% to needs: rent, groceries, utilities.\n"
        "30% to wants: dining out, entertainment, hobbies.\n"
        "20% to savings and debt repayment.\n"
    ),
    "march_2024_transactions.txt": (
        "March 2024 Transaction Summary\n"
        "Groceries: $520 across 8 transactions\n"
        "Utilities: $180 (electricity $90, water $50, internet $40)\n"
        "Entertainment: $95 (streaming $35, cinema $60)\n"
        "Dining out: $210 across 12 transactions\n"
        "Transport: $145 (fuel $95, public transit $50)\n"
    ),
}

for filename, content in docs.items():
    (DOCS_DIR / filename).write_text(content)

print("✅ Documents created in 'financial_docs/'")

# ============================================================
# CELL 4 — FinancialRAG class
# ============================================================
class FinancialRAG:

    def __init__(self, docs_dir="financial_docs", top_k=3):
        self.docs_dir = Path(docs_dir)
        self.top_k = top_k
        self.chunks = []
        self._load()

    def _load(self):
        self.chunks = []
        for path in self.docs_dir.glob("*.txt"):
            text = path.read_text()
            words = text.split()
            size, overlap = 200, 40
            for i in range(0, len(words), size - overlap):
                chunk = " ".join(words[i : i + size])
                self.chunks.append({"source": path.name, "text": chunk})
        logger.info("Loaded %d chunks", len(self.chunks))

    def _retrieve(self, query):
        query_words = set(query.lower().split())
        scored = [(len(query_words & set(c["text"].lower().split())), c) for c in self.chunks]
        scored.sort(key=lambda x: x[0], reverse=True)
        results = [c for score, c in scored[:self.top_k] if score > 0]
        return results or [self.chunks[0]]

    def ask(self, query, verbose=False):
        query = query.strip()
        if not query:
            raise ValueError("Query cannot be empty.")

        context = "\n\n---\n\n".join(
            "[" + c["source"] + "]\n" + c["text"] for c in self._retrieve(query)
        )

        if verbose:
            logger.info("Context:\n%s", context)

        system_prompt = (
            "You are a helpful personal finance assistant. "
            "Answer using ONLY the context below. "
            "If the answer is not in the context, say so.\n\n"
            "CONTEXT:\n" + context
        )

        try:
            response = client.chat.completions.create(
                model=MODEL,
                messages=[
                    {"role": "system", "content": system_prompt},
                    {"role": "user", "content": query},
                ],
                temperature=0.2,
                max_tokens=512,
            )
            return response.choices[0].message.content.strip()
        except Exception as exc:
            logger.error("LLM call failed: %s", exc)
            raise RuntimeError("LLM call failed: " + str(exc)) from exc

    def interactive(self):
        print("\n💬 Financial RAG — type 'exit' to quit\n")
        while True:
            try:
                query = input("You: ").strip()
            except (EOFError, KeyboardInterrupt):
                print("\nSession ended.")
                break
            if query.lower() in {"exit", "quit", ""}:
                print("Goodbye!")
                break
            try:
                print("\nAssistant: " + self.ask(query) + "\n")
            except Exception as exc:
                print("Error: " + str(exc))

finance_rag = FinancialRAG()
print("✅ FinancialRAG ready")

# ============================================================
# CELL 5 — Sample questions
# ============================================================
questions = [
    "What were the top spending categories in Q1 2024?",
    "How much did I spend on dining out in March 2024?",
    "Explain the 50/30/20 budgeting rule.",
    "What was my net savings in Q1 2024?",
]

for q in questions:
    print("Q: " + q)
    print("A: " + finance_rag.ask(q))
    print("-" * 60)

# ============================================================
# CELL 6 — Interactive chat (uncomment to use)
# ============================================================
finance_rag.interactive()
````

````python
# ============================================================
# CELL 1 — Install
# ============================================================
!pip install pandas openai --quiet
print("Packages ready")

# ============================================================
# CELL 2 — Imports & LLM config
# ============================================================
import os, io, math, time, logging
import pandas as pd
from openai import OpenAI

client = OpenAI(
    api_key  = "Curtin2026ISYS20015002",
    base_url = "https://ollama.locollm.org/v1",
)
MODEL = "gemma3:4b"

logging.basicConfig(level=logging.INFO, format="%(asctime)s [%(levelname)s] %(message)s")
logger = logging.getLogger("integration_tests")
print("Config ready")

# ============================================================
# CELL 3 — All core functions (self-contained)
# ============================================================

def load_and_clean_csv(source):
    if isinstance(source, str) and "\n" in source:
        df = pd.read_csv(io.StringIO(source))
    else:
        df = pd.read_csv(source)
    required = {"Category", "Amount"}
    if not required.issubset(df.columns):
        raise ValueError("Missing columns: " + str(required - set(df.columns)))
    df["Amount"] = (
        df["Amount"].astype(str)
        .str.replace("$", "", regex=False)
        .str.replace(",", "", regex=False)
        .str.strip()
    )
    df["Amount"] = pd.to_numeric(df["Amount"], errors="coerce")
    df = df.dropna(subset=["Amount"])
    return df


def analyze_spending_patterns(df):
    if df.empty:
        return {"error": "No data to analyse."}
    category_totals      = df.groupby("Category")["Amount"].sum().sort_values(ascending=False)
    total_spending       = round(float(category_totals.sum()), 2)
    category_percentages = (category_totals / total_spending * 100).round(2)
    top_category         = str(category_totals.idxmax())
    top_percentage       = float(category_percentages.max())
    insights = ["Top category: " + top_category + " (" + str(top_percentage) + "% of total)"]
    if top_percentage > 40:
        insights.append("Warning: " + top_category + " exceeds 40% of total spending.")
    if len(category_totals) >= 3:
        bottom = str(category_totals.idxmin())
        insights.append("Lowest: " + bottom + " (" + str(float(category_percentages.min())) + "%)")
    return {
        "Total Spending":       total_spending,
        "Category Totals":      {str(k): round(float(v), 2) for k, v in category_totals.items()},
        "Category Percentages": {str(k): round(float(v), 2) for k, v in category_percentages.items()},
        "Insights":             insights,
    }


def savings_calculator(current_savings, monthly_contribution, target_amount, annual_rate=0.0):
    try:
        current_savings      = float(current_savings)
        monthly_contribution = float(monthly_contribution)
        target_amount        = float(target_amount)
        annual_rate          = float(annual_rate)
    except (TypeError, ValueError):
        return {"error": "All inputs must be numeric."}
    if target_amount <= 0 or monthly_contribution <= 0:
        return {"error": "Target and contribution must be > 0."}
    if current_savings < 0:
        return {"error": "Savings cannot be negative."}
    if current_savings >= target_amount:
        return {"months_needed": 0, "years_needed": 0, "message": "Goal already achieved."}
    monthly_rate = annual_rate / 100 / 12
    if monthly_rate > 0:
        balance, months_needed = current_savings, 0
        while balance < target_amount:
            balance = balance * (1 + monthly_rate) + monthly_contribution
            months_needed += 1
            if months_needed > 12000:
                return {"error": "Goal unreachable with current parameters."}
    else:
        months_needed = math.ceil((target_amount - current_savings) / monthly_contribution)
    return {
        "months_needed": months_needed,
        "years_needed":  months_needed // 12,
        "extra_months":  months_needed % 12,
    }


def chat_with_assistant(message, history=None):
    if history is None:
        history = []
    if not message.strip():
        return "Please enter a message.", history
    system_prompt = (
        "You are a helpful personal finance assistant. "
        "Give clear, concise, practical advice about budgeting and saving."
    )
    messages = [{"role": "system", "content": system_prompt}]
    for user_msg, bot_msg in history:
        messages.append({"role": "user",      "content": user_msg})
        messages.append({"role": "assistant", "content": bot_msg})
    messages.append({"role": "user", "content": message})
    try:
        response = client.chat.completions.create(
            model       = MODEL,
            messages    = messages,
            temperature = 0.3,
            max_tokens  = 256,
        )
        reply = response.choices[0].message.content.strip()
        history.append((message, reply))
        return reply, history
    except Exception as e:
        return "LLM error: " + str(e), history

print("Core functions defined")

# ============================================================
# CELL 4 — Test result tracker
# ============================================================

class TestRunner:
    def __init__(self, suite_name):
        self.suite_name = suite_name
        self.passed     = 0
        self.failed     = 0
        self.errors     = []

    def check(self, name, condition, detail=""):
        if condition:
            print("  PASS: " + name)
            self.passed += 1
        else:
            msg = "  FAIL: " + name + (" | " + detail if detail else "")
            print(msg)
            self.errors.append(msg)
            self.failed += 1

    def summary(self):
        total = self.passed + self.failed
        print(
            "\n" + self.suite_name + " results: "
            + str(self.passed) + "/" + str(total) + " passed"
        )
        return self.failed == 0

# ============================================================
# CELL 5 — Integration test: full workflow
# ============================================================

def test_full_workflow():
    print("\n" + "="*50)
    print("TEST 1: Full Workflow Integration")
    print("="*50)
    t = TestRunner("Full Workflow")

    # --- Step 1: Load CSV ---
    sample_csv = (
        "Category,Amount\n"
        "Rent,1800.00\n"
        "Woolworths,320.50\n"
        "Coles,180.30\n"
        "AGL Energy,145.00\n"
        "Telstra,89.00\n"
        "Uber,55.20\n"
        "Event Cinemas,42.00\n"
        "Gym,65.00\n"
    )
    print("\nStep 1: Load CSV")
    try:
        df = load_and_clean_csv(sample_csv)
        t.check("CSV loaded successfully",   not df.empty)
        t.check("Correct row count",         len(df) == 8, "expected 8, got " + str(len(df)))
        t.check("Amount column is numeric",  pd.api.types.is_numeric_dtype(df["Amount"]))
    except Exception as e:
        t.check("CSV load did not crash", False, str(e))
        return t.summary()

    # --- Step 2: Run analysis pipeline ---
    print("\nStep 2: Analysis pipeline")
    try:
        result = analyze_spending_patterns(df)
        t.check("Analysis returns dict",         isinstance(result, dict))
        t.check("No error in result",            "error" not in result)
        t.check("Total Spending present",        "Total Spending" in result)
        t.check("Category Totals present",       "Category Totals" in result)
        t.check("Category Percentages present",  "Category Percentages" in result)
        t.check("Insights present",              "Insights" in result and len(result["Insights"]) > 0)
        pct_sum = round(sum(result["Category Percentages"].values()), 1)
        t.check("Percentages sum to 100",        pct_sum == 100.0, "got " + str(pct_sum))
        top = list(result["Category Totals"].keys())[0]
        t.check("Rent is top category",          top == "Rent", "got " + top)
    except Exception as e:
        t.check("Analysis did not crash", False, str(e))

    # --- Step 3: Savings tool ---
    print("\nStep 3: Savings calculator tool")
    try:
        total_spent = result.get("Total Spending", 2700)
        goal        = round(total_spent * 3, 2)
        calc_result = savings_calculator(
            current_savings      = 500,
            monthly_contribution = 300,
            target_amount        = goal,
        )
        t.check("Calculator returns dict",       isinstance(calc_result, dict))
        t.check("No error from calculator",      "error" not in calc_result)
        t.check("months_needed > 0",             calc_result.get("months_needed", 0) > 0)
        t.check("years_needed is int",           isinstance(calc_result.get("years_needed"), int))
    except Exception as e:
        t.check("Calculator did not crash", False, str(e))

    # --- Step 4: Chat response about the data ---
    print("\nStep 4: Chat response")
    try:
        top_cat     = list(result["Category Totals"].keys())[0]
        top_amt     = result["Category Totals"][top_cat]
        question    = (
            "My top spending category is " + top_cat +
            " at $" + str(top_amt) + ". Give me one tip to reduce it."
        )
        reply, history = chat_with_assistant(question)
        t.check("Chat returns a string",         isinstance(reply, str))
        t.check("Reply is not empty",            len(reply.strip()) > 0)
        t.check("Reply is not an error",         not reply.startswith("LLM error"))
        t.check("History updated",               len(history) == 1)
    except Exception as e:
        t.check("Chat did not crash", False, str(e))

    # --- Step 5: Multi-turn conversation ---
    print("\nStep 5: Multi-turn conversation")
    try:
        reply2, history2 = chat_with_assistant(
            "What is the 50/30/20 rule?", history=[]
        )
        reply3, history3 = chat_with_assistant(
            "How does that apply to my situation?", history=history2
        )
        t.check("Second turn returns string",    isinstance(reply3, str))
        t.check("History grows correctly",       len(history3) == 2)
        t.check("Follow-up reply not empty",     len(reply3.strip()) > 0)
    except Exception as e:
        t.check("Multi-turn did not crash", False, str(e))

    return t.summary()

# ============================================================
# CELL 6 — Integration test: error handling
# ============================================================

def test_error_handling():
    print("\n" + "="*50)
    print("TEST 2: Error Handling")
    print("="*50)
    t = TestRunner("Error Handling")

    # --- Invalid file uploads ---
    print("\nInvalid file inputs")

    # Wrong columns
    try:
        load_and_clean_csv("Name,Value\nWoolworths,100\n")
        t.check("Wrong columns raises error", False, "no error raised")
    except ValueError as e:
        t.check("Wrong columns raises ValueError", True)

    # Completely empty string
    try:
        df = load_and_clean_csv("Category,Amount\n")
        t.check("Empty CSV returns empty df", df.empty)
    except Exception as e:
        t.check("Empty CSV handled", False, str(e))

    # All invalid amounts
    try:
        df = load_and_clean_csv("Category,Amount\nRent,abc\nFood,xyz\n")
        t.check("All-invalid amounts returns empty df", df.empty)
    except Exception as e:
        t.check("All-invalid amounts handled", False, str(e))

    # --- Malformed data ---
    print("\nMalformed data")

    # Dollar signs
    try:
        df = load_and_clean_csv("Category,Amount\nRent,$1800\nFood,$320\n")
        t.check("Dollar signs cleaned",  df["Amount"].iloc[0] == 1800.0)
    except Exception as e:
        t.check("Dollar signs handled", False, str(e))

    # Comma-formatted numbers
    try:
        df = load_and_clean_csv("Category,Amount\nRent,\"1,800\"\nFood,320\n")
        t.check("Comma numbers cleaned", df["Amount"].iloc[0] == 1800.0)
    except Exception as e:
        t.check("Comma numbers handled", False, str(e))

    # Mixed valid and invalid rows
    try:
        df = load_and_clean_csv("Category,Amount\nRent,1800\nFood,bad\nUber,55\n")
        t.check("Mixed rows: valid kept",   len(df) == 2, "expected 2, got " + str(len(df)))
        t.check("Mixed rows: no NaN left",  df["Amount"].isna().sum() == 0)
    except Exception as e:
        t.check("Mixed rows handled", False, str(e))

    # --- Savings calculator edge cases ---
    print("\nSavings calculator edge cases")

    edge_cases = [
        ("Zero contribution",   dict(current_savings=0,    monthly_contribution=0,    target_amount=5000)),
        ("Zero target",         dict(current_savings=0,    monthly_contribution=500,  target_amount=0)),
        ("Negative savings",    dict(current_savings=-100, monthly_contribution=500,  target_amount=5000)),
        ("Negative rate",       dict(current_savings=0,    monthly_contribution=500,  target_amount=5000, annual_rate=-1)),
        ("String inputs",       dict(current_savings="abc",monthly_contribution="xyz",target_amount="???")),
    ]

    for label, kwargs in edge_cases:
        try:
            result = savings_calculator(**kwargs)
            t.check(label + " returns error key", "error" in result,
                    "got: " + str(result))
        except Exception as e:
            t.check(label + " handled gracefully", False, str(e))

    # --- Analysis edge cases ---
    print("\nAnalysis edge cases")

    # Single row
    try:
        df = load_and_clean_csv("Category,Amount\nRent,1800\n")
        result = analyze_spending_patterns(df)
        t.check("Single row: no crash",       "error" not in result)
        t.check("Single row: 100% pct",       list(result["Category Percentages"].values())[0] == 100.0)
    except Exception as e:
        t.check("Single row handled", False, str(e))

    # All same category
    try:
        df = load_and_clean_csv("Category,Amount\nFood,100\nFood,200\nFood,300\n")
        result = analyze_spending_patterns(df)
        t.check("Duplicate categories merged", result["Category Totals"].get("Food") == 600.0)
    except Exception as e:
        t.check("Duplicate categories handled", False, str(e))

    # --- Chat input edge cases ---
    print("\nChat input edge cases")

    empty_reply, _ = chat_with_assistant("   ")
    t.check("Empty message handled",     isinstance(empty_reply, str) and len(empty_reply) > 0)

    long_msg  = "Tell me about budgeting. " * 50
    long_reply, _ = chat_with_assistant(long_msg)
    t.check("Very long message handled", isinstance(long_reply, str) and len(long_reply) > 0)

    return t.summary()

# ============================================================
# CELL 7 — Performance tests
# ============================================================

def test_performance():
    print("\n" + "="*50)
    print("TEST 3: Performance")
    print("="*50)
    t = TestRunner("Performance")

    # Large CSV (500 rows)
    print("\nLarge dataset (500 rows)")
    rows = ["Category,Amount"]
    cats = ["Rent", "Food", "Transport", "Utilities", "Entertainment"]
    for i in range(500):
        rows.append(cats[i % len(cats)] + "," + str(round(50 + i * 0.5, 2)))
    large_csv = "\n".join(rows)

    start = time.time()
    try:
        df     = load_and_clean_csv(large_csv)
        result = analyze_spending_patterns(df)
        elapsed = time.time() - start
        t.check("500 rows loads correctly",     len(df) == 500)
        t.check("500 rows analysis correct",    "error" not in result)
        t.check("500 rows under 2 seconds",     elapsed < 2.0,
                "took " + str(round(elapsed, 2)) + "s")
    except Exception as e:
        t.check("Large dataset handled", False, str(e))

    # Repeated calculator calls
    print("\nRepeated calculator calls (100x)")
    start = time.time()
    errors = 0
    for i in range(100):
        r = savings_calculator(i * 10, 500, 10000)
        if "error" in r and i > 0:
            errors += 1
    elapsed = time.time() - start
    t.check("100 calculator calls succeed",   errors == 0, str(errors) + " errors")
    t.check("100 calls under 1 second",       elapsed < 1.0,
            "took " + str(round(elapsed, 2)) + "s")

    return t.summary()

# ============================================================
# CELL 8 — Run all integration tests
# ============================================================
print("INTEGRATION TEST SUITE")
print("=" * 50)

suite_results = []
suite_results.append(test_full_workflow())
suite_results.append(test_error_handling())
suite_results.append(test_performance())

print("\n" + "=" * 50)
passed_suites = sum(suite_results)
total_suites  = len(suite_results)
print("FINAL: " + str(passed_suites) + "/" + str(total_suites) + " suites passed")
if all(suite_results):
    print("ALL INTEGRATION TESTS PASSED")
else:
    print("Some tests failed - review output above for details")
````
